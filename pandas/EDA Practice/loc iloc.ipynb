# loc and iloc for PANDAS In selection Data

### 1.loc — Label-based Indexing

loc tab use hota hai jab aap row ya column ka naam (label) use kar ke data select karna chahte hain.

Yani: Row/Column ka NAAM likhte hain

### 2. iloc — Position-based Indexing

iloc tab use hota hai jab aap row or column ki numeric position se data select karte hain.

Yani: Row numbers (0,1,2...) use hote hain
